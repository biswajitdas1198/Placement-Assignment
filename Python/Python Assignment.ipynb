## Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might be repeated characters in the string. Your task is to find the highest frequency and returns the length of the highest-frequency word.

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same. Example input - string = “write write write all the number from from from 1 to 100” Example output - 5 Explanation - From the given string we can note that the most frequent words are “write” and “from” and the maximum value of both the values is “write” and its corresponding length is 5

```python
def find_highest_frequency_word_length(string):
    words = string.split()

    word_frequency = {}
    for word in words:
        word_frequency[word] = word_frequency.get(word, 0) + 1

    max_frequency = max(word_frequency.values())

    highest_frequency_word = max(word_frequency, key=word_frequency.get)
    highest_frequency_word_length = len(highest_frequency_word)

    return highest_frequency_word_length


# Test case 1
string1 = "write write write all the number from from from 1 to 100"
output1 = find_highest_frequency_word_length(string1)
print(output1)  # Output: 5

# Test case 2
string2 = "Sweet home home home"
output2 = find_highest_frequency_word_length(string2)
print(output2)  # Output: 4

# Test case 3
string3 = "apple apple apple orange orange orange orange orange"
output3 = find_highest_frequency_word_length(string3)
print(output3)  # Output: 6
```

## Question 4 -
Write a program to download the data from the link given below and then read the data and convert the into the proper structure and return it as a CSV file. Link - https://data.nasa.gov/resource/y77d-th95.json Note - Write code comments wherever needed for code understanding.

Sample Data -

Excepted Output Data Attributes

    ● Name of Earth Meteorite - string id - ID of Earth 
    ● Meteorite - int nametype - string recclass - string 
    ● mass - Mass of Earth Meteorite - float year - Year at which Earth 
    ● Meteorite was hit - datetime format reclat - float recclong - float 
    ● point coordinates - list of int

```python
import requests
import pandas as pd

# Function to download data from a URL
def download_data(url):
    response = requests.get(url)
    data = response.json()
    return data

# Function to process the data
def process_data(data):
    # Create an empty list to store the processed data
    processed_data = []

    # Iterate over each record in the data
    for record in data:
        # Extract the desired fields from each record and append them to the processed data list
        processed_data.append({
            'name': record.get('name', ''),
            'id': record.get('id', ''),
            'nametype': record.get('nametype', ''),
            'recclass': record.get('recclass', ''),
            'mass': record.get('mass (g)', ''),
            'year': pd.to_datetime(record.get('year', ''), errors='coerce'),
            'reclat': record.get('reclat', ''),
            'reclong': record.get('reclong', '')
        })

    return processed_data

# Function to save the processed data as a CSV file
def save_as_csv(data, filename):
    # Convert the processed data into a DataFrame
    df = pd.DataFrame(data)

    # Save the DataFrame to a CSV file
    df.to_csv(filename, index=False)

# URL to download the data from
data_url = 'https://data.nasa.gov/resource/y77d-th95.json'

# Download the data
nasa_data = download_data(data_url)

# Process the data
processed_data = process_data(nasa_data)

# Save the processed data as a CSV file
output_filename = 'nasa_data.csv'
save_as_csv(processed_data, output_filename)

print('Data downloaded, processed, and saved as', output_filename)
```

## Question 2: -
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

    Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
    Example output 1- YES
    Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
    character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
    Example output 2 - NO

```python
from collections import Counter

def is_valid_string(s):
    char_freq = Counter(s)
    freq_freq = Counter(char_freq.values())

    if len(freq_freq) == 1:
        return "YES"
    elif len(freq_freq) == 2:
        if freq_freq[1] == 1 or (freq_freq[min(freq_freq)] == 1 and abs(min(freq_freq) - max(freq_freq)) == 1):
            return "YES"
    return "NO"

# Example test cases
print(is_valid_string("abc"))  # Output: YES
print(is_valid_string("abcc"))  # Output: NO

# Additional test cases
print(is_valid_string("aabbbccc"))  # Output: YES
print(is_valid_string("aabbccd"))  # Output: YES
print(is_valid_string("aaaabbbbcc"))  # Output: NO
```

## Question 7 -
Using the data from Question 4, write code to analyze the data and answer the following questions Note -
1. Draw plots to demonstrate the analysis for the following questions for better visualizations
2. Write code comments wherever required for code understanding

Insights to be drawn -
● Get all the Earth meteorites that fell before the year 2000
● Get all the earth meteorites co-ordinates who fell before the year 1970
● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more
than 10000kg

```python
import requests
import matplotlib.pyplot as plt

# Fetch the data from the NASA API
url = "https://data.nasa.gov/resource/y77d-th95.json"
response = requests.get(url)
data = response.json()

# Prepare lists to store relevant data
meteorites_before_2000 = []
meteorites_before_1970_coordinates = []
meteorites_mass_over_10000kg = []

# Iterate over the meteorite data
for meteorite in data:
    # Check if the meteorite fell before the year 2000
    year = meteorite.get("year")
    if year and int(year.split("-")[0]) < 2000:
        meteorites_before_2000.append(meteorite)
        
        # Check if the meteorite fell before the year 1970 and has coordinates
        if year and int(year.split("-")[0]) < 1970 and meteorite.get("reclat") and meteorite.get("reclong"):
            meteorites_before_1970_coordinates.append((float(meteorite["reclat"]), float(meteorite["reclong"])))
        
        # Check if the meteorite has a mass greater than 10000kg
        mass = meteorite.get("mass (g)")
        if mass and int(mass) >= 10000:
            meteorites_mass_over_10000kg.append(meteorite)

# Plotting the data

# Plot 1: Meteorites fell before the year 2000
years = [int(meteorite.get("year").split("-")[0]) for meteorite in meteorites_before_2000]
plt.hist(years, bins=range(1800, 2001, 10), edgecolor="black")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.title("Meteorites Fell before 2000")
plt.show()

# Plot 2: Meteorites' coordinates that fell before the year 1970
latitudes = [coord[0] for coord in meteorites_before_1970_coordinates]
longitudes = [coord[1] for coord in meteorites_before_1970_coordinates]
plt.scatter(longitudes, latitudes, color="red", marker="o", alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Meteorites' Coordinates Fell before 1970")
plt.show()

# Plot 3: Meteorites with mass over 10000kg
masses = [int(meteorite.get("mass (g)")) / 1000 for meteorite in meteorites_mass_over_10000kg]  # Convert grams to kg
plt.boxplot(masses)
plt.ylabel("Mass (kg)")
plt.title("Meteorites with Mass over 10000kg")
plt.show()

# Additional insights

# Total number of meteorites fell before 2000
num_meteorites_before_2000 = len(meteorites_before_2000)
print(f"Number of meteorites fell before 2000: {num_meteorites_before_2000}")

# Number of meteorites' coordinates fell before 1970
num_coordinates_before_1970 = len(meteorites_before_1970_coordinates)
print(f"Number of meteorites' coordinates fell before 1970: {num_coordinates_before_1970}")

# Number of meteorites with mass over 10000kg
num_mass_over_10000kg = len(meteorites_mass_over_10000kg)
print(f"Number of meteorites with mass over 10000kg: {num_mass_over_10000kg}")
```

## Question 8 -
Using the data from Question 5, write code the analyze the data and answer the following questions Note -
1. Draw plots to demonstrate the analysis for the following questions and better visualizations
2. Write code comments wherever required for code understanding

Insights to be drawn -
● Get all the overall ratings for each season and using plots compare the ratings for all the
seasons, like season 1 ratings, season 2, and so on.
● Get all the episode names, whose average rating is more than 8 for every season
● Get all the episode names that aired before May 2019
● Get the episode name from each season with the highest and lowest rating
● Get the summary for the most popular ( ratings ) episode in every season

```python
import requests
import matplotlib.pyplot as plt

# Fetch the data from the TVMaze API
url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
response = requests.get(url)
data = response.json()

# Get all the episodes
episodes = data["_embedded"]["episodes"]

# Prepare dictionaries to store the ratings and episode information
season_ratings = {}
episode_ratings = {}
episode_names_above_8 = {}
episode_names_before_may_2019 = {}
highest_rated_episodes = {}
lowest_rated_episodes = {}
most_popular_episodes_summary = {}

# Iterate over the episodes
for episode in episodes:
    season = episode["season"]
    rating = episode["rating"]["average"]
    name = episode["name"]
    air_date = episode["airdate"]

    # Store the overall ratings for each season
    if season in season_ratings:
        season_ratings[season].append(rating)
    else:
        season_ratings[season] = [rating]

    # Store the episode ratings for each episode
    episode_ratings[name] = rating

    # Check if the average rating is above 8 for each season
    if rating > 8:
        if season in episode_names_above_8:
            episode_names_above_8[season].append(name)
        else:
            episode_names_above_8[season] = [name]

    # Check if the episode aired before May 2019
    if air_date < "2019-05-01":
        if season in episode_names_before_may_2019:
            episode_names_before_may_2019[season].append(name)
        else:
            episode_names_before_may_2019[season] = [name]

    # Store the highest and lowest rated episodes for each season
    if season in highest_rated_episodes:
        if rating > episode_ratings[highest_rated_episodes[season]]:
            highest_rated_episodes[season] = name
        if rating < episode_ratings[lowest_rated_episodes[season]]:
            lowest_rated_episodes[season] = name
    else:
        highest_rated_episodes[season] = name
        lowest_rated_episodes[season] = name

    # Store the summary for the most popular episode in each season
    if season in most_popular_episodes_summary:
        if rating > episode_ratings[most_popular_episodes_summary[season]]:
            most_popular_episodes_summary[season] = name
    else:
        most_popular_episodes_summary[season] = name

# Plotting the data

# Plot 1: Ratings comparison for all seasons
seasons = list(season_ratings.keys())
ratings = [sum(season_ratings[season]) / len(season_ratings[season]) for season in seasons]
plt.bar(seasons, ratings)
plt.xlabel("Season")
plt.ylabel("Average Rating")
plt.title("Ratings Comparison for All Seasons")
plt.show()

# Additional insights

# Print overall ratings for each season
for season in seasons:
    print(f"Season {season} Average Rating: {sum(season_ratings[season]) / len(season_ratings[season])}")

# Print episode names with average rating above 8 for each season
for season in episode_names_above_8:
    print(f"Season {season} Episodes with Rating Above 8: {', '.join(episode_names_above_8[season])}")

# Print episode names that aired before May 2019 for each season
for season in episode_names_before_may_2019:
    print(f"Season {season} Episodes Aired Before May 2019: {', '.join(episode_names_before_may_2019[season])}")

# Print the episode with the highest and lowest rating for each season
for season in highest_rated_episodes:
    print(f"Season {season} Highest Rated Episode: {highest_rated_episodes[season]}")
    print(f"Season {season} Lowest Rated Episode: {lowest_rated_episodes[season]}")

# Print the summary for the most popular episode in each season
for season in most_popular_episodes_summary:
    print(f"Season {season} Most Popular Episode: {most_popular_episodes_summary[season]}")
```

## Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    
    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)
    
    # Initialize counts
    counts = {'Noun': 0, 'Verb': 0, 'Pronoun': 0, 'Adjective': 0}
    
    # Count the number of each POS tag
    for word, tag in tagged_words:
        if tag.startswith('N'):
            counts['Noun'] += 1
        elif tag.startswith('V'):
            counts['Verb'] += 1
        elif tag.startswith('PRP'):
            counts['Pronoun'] += 1
        elif tag.startswith('JJ'):
            counts['Adjective'] += 1
    
    return counts

# Test cases
text1 = "The quick brown fox jumps over the lazy dog."
text2 = "I like to eat pizza on Fridays. It is delicious."

# Count POS tags in text1
counts1 = count_pos_tags(text1)
print("POS counts for text1:", counts1)

# Count POS tags in text2
counts2 = count_pos_tags(text2)
print("POS counts for text2:", counts2)
```
